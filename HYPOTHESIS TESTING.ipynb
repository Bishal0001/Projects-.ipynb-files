````python
w0 = df[df['workingday'] == 0][['workingday','count']]
w1 = df[df['workingday'] == 1][['workingday','count']]
w1m = w1['count']
w0m = w0['count']
````

````python
log_w0m = np.log10(w0m)
log_w1m = np.log10(w1m)

boxcox_w0m = stats.boxcox(w0m)
boxcox_w0m = boxcox_w0m[0]

boxcox_w1m = stats.boxcox(w1m)
boxcox_w1m = boxcox_w1m[0]

````

# Chi-Sq Test

````python
chi_tab = pd.crosstab(df.weather,df.season,values=df['count'],aggfunc=sum)
chi_tab
````

````python
stat, p_value, dof, expected = stats.chi2_contingency(chi_tab.loc[1:3])
stat, p_value, dof
print(f'T Statistics and P Value of above Chi Sq Test is {stat} and {p_value}.')
````

```As chi sq p value  < alpha, we fail to accept null hypothesis```
